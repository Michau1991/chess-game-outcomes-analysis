This analysis is based on a public Lichess dataset and focuses on rated games only

## Project objective

The goal of this project is to analyze how online chess games are lost and whether the way of losing depends on player rating.

Main research question:
- Doest the way a chess game ends depend on player skill level?

Sub-questions explored in this notebook:
- How are games most commonly decided?
- Do stronger players resign more often than they get checkmated?

## Data cleaning and preparation

To ensure meaningful and consisten analysis, several preprocessing steps were applied:

- Only rated games were selected to focus on competitive play and reduce noise from casual matches.
- Draws were excluded from the main outcome analysis to focus on decisive results (resignation, checkmate, timeout).
- A new variable "avg_rating" was created as the average of both players' ratings to represent overall game skills level.
- Games were grouped into rating categories to allow comparison across skill levels.
- The value "outoftime" was renamed to "timeout" for improved readability.

Missing values were checked only for variables used in the analysis, and no missing values were found. 

These steps ensure that the dataset is clean, consisten, and suitable for further analysis.

```python
df_clean = df[df["rated"]==True].copy()

df_clean["victory_status"] = df_clean["victory_status"].replace({"outoftime": "timeout"
                                                                })

df_clean.shape
```

```python
main_statuses = ["resign", "mate", "timeout"]
df_main = df_clean[df_clean["victory_status"].isin(main_statuses)].copy()
df_main["victory_status"].value_counts(normalize=True) * 100
```

```python
df_main["avg_rating"] = (df_main["white_rating"] + df_main["black_rating"])/2
```

```python
bins = [0,1200,1600,2000, np.inf]
labels = ["<1200", "1200-1599", "1600-1999", "2000+"]

df_main["rating_group"] = pd.cut(
    df_main['avg_rating'],
    bins=bins,
    labels=labels,
    right=False
)

```

```python
df_main["rating_group"].value_counts().sort_index()
```

## Research Question 1
### How do rated games most commonly end?

The first step is to examine the overall distribution of game outcomes in rated games.
This provides context for the rest of the analysis.

This provides a complete overview of how games typically end. In the next steps, draws are excluded to focus on decisive outcomes (resignation, checkmate, timeout).

```python
df_clean["victory_status"].value_counts()
```

```python
plt.figure(figsize=(8,5))

sns.countplot(
    data=df_clean,
    x="victory_status",
    order=df_clean["victory_status"].value_counts().index
)

plt.title("Distribution of game outcomes")
plt.xlabel("Victory status")
plt.ylabel("Number of games")

plt.show()
```

### Interpretation

Resignation is the most common way games end, significally more frequent than checkmate or timeout.
This suggests that many games are decided before checkmate occurs, as players tend to resign once a losing position becomes clear.
Checkmate remains relatively common, while timeout represents a smaller portion of outcomes.
Draws are present but form a minority of games, which justifies focusing further analysis on decisive outcomes.

## Research Question 2
### Does game outcome depend on player rating?

To explore wheter player skill influences how games are lost, decisive outcomes (resignation, checkmate, timeout) are compared across different rating groups.

```python
rating_vs_status = pd.crosstab(
    df_main["rating_group"],
    df_main["victory_status"],
    normalize="index"
) *100

```

```python
rating_vs_status.round(2)
```

```python
rating_vs_status.plot(kind="bar", figsize=(10, 6))

plt.title("Game outcome vs player rating")
plt.xlabel("Average rating")
plt.ylabel("Percentage of games")
plt.xticks(rotation=0)

plt.legend(title="Outcome")

plt.show()
```

### Interpretation

As player rating increases, the proportion of games ending in resignation rises significantly, while the proportion of checkmates decreases.

This suggest that stronger players are more likely to recognize losing positions and resign instead of playing until checkmate.

In contrast, lower-rated players tend to continue playing in losing positions, resulting in a higher frequency of checkmates.

Timeout frequency appears relatively stable across rating groups and does not show strong dependency on player skill level.

## Research Question 3
### Which opening types are most frequently chosen?

The opening phase sets the direction of a chess game, influencing the type of positions that arise later on. Diffrent openings lead to diffrent styles of play, rangin from dynamic and tactical to slower paced and more strategic.
To better understand these patterns, games are grouped using ECO codes.
ECO (Encyclopedia of Chess Openings) is a standarized system that classifies chess openings into five broad categories, labeled from A to E.
Instead of analyzing specific opening names, the analysis focuses on broader groups defined by the first letter of the ECO code, which allows for more meaningful insights.



```python
df_clean["opening_group"] = df_clean["opening_eco"].str[0]
```

```python
df_clean["opening_group"].value_counts()
```

```python
plt.figure(figsize=(8,5))

sns.countplot(
    data=df_clean,
    x="opening_group",
    order=df_clean["opening_group"].value_counts().index
)

plt.title("Distribution of opening types")
plt.xlabel("Opening group (ECO)")
plt.ylabel("Number of games")

plt.show()
```

### Interpretation

Opening choices are concentrated in a few groups, with categories C and B making up  the majority of games.
These groups are typically associated with open and semi-open games, which tend to lead to more dynamic positions with faster piece activity and earlier tactical opportunities.
In contrast, closed and semi-closed openings (groups D and E) appear less frequently and are generally linked to a slower, more strategic play, where positions develop more gradually.
Group A represents a more diverse set of opening systems that do not fit as clearly into the typical open or closed structures, which may explain its moderate but consistent presence.
Overall, this suggests that most games tend toward more dynamic and active play, while slower and more positional approaches appear less frequently.

## Research Question 4
### Do opening choices differ by player rating?

Opening choice is often a matter of player preference and playing style, influencing the type of positions that arise in a game.
This analysis explores whether these preferences can be linked to player rating, and whether players at diffrent skill levels tend to favor different types of openings.

```python
df_clean["avg_rating"] = (df_clean["white_rating"] + df_clean["black_rating"])/2

bins = [0,1200,1600,2000, np.inf]
labels = ["<1200", "1200-1599", "1600-1999", "2000+"]

df_clean["rating_group"] = pd.cut(
    df_clean["avg_rating"],
    bins=bins,
    labels=labels,
    right=False
)
```

```python
opening_vs_rating = pd.crosstab(
    df_clean["rating_group"],
    df_clean["opening_group"],
    normalize="index"
) *100

```

```python
opening_vs_rating.plot(kind="bar", figsize=(10,6))

plt.title("Opening type vs player rating")
plt.xlabel("Average rating")
plt.ylabel("Percentage of games")
plt.xticks(rotation=0)
plt.legend(title="Opening group", bbox_to_anchor=(1.05, 1), loc="upper left")

plt.show()
```

### Interpretation

Opening preferences vary noticeably across rating levels, indicating that players approach the game diffrently depending on their skill.
Lower-rated players tend to favor opening groups associated with more direct and dynamic play, where early activiy and tactical opportunities are more common.
As rating increases, the distribution of opening choices becomes more balanced, with a growing share of closed and semi-closed structures that typically require more strategic plannign and positional understanding.
At the same time, the dominance of open games decreases, suggesting that stronger players are less reliant on immediate tactical play and more comfortable navigating a wider range of positions.
Overall, higher-rated players appear to use a more diverse set of opening strategies, reflecting a broader understanding of different types of positions.
This may suggest that lower-rated players tend to choose openings based on simplcity and accessibility, while higher-rated players are more likely to select openings aligned with their preferences, supported by greater experience and a deeper understanding of strategic concepts.

## Final Conclusions
This analysis shows that the way chess games unfold is strongly connected to player behavior and skill level.
Most games end by resignation rather than checkmate, suggesting that players often recognize losing positions before the game reaches its final stage.
Game outcomes and opening choices both vary across rating levels, indicating that stronger players approach the game diffrently from the very beginning.
Higher-rated players tend to use wider range of openings and are more likely to choose structures associated with strategic and positional play, while lower-rated players rely more on direct and dynamic setups.
Overall, the results hightlight how player decisions - from the opening moves to the final outcome - evolve with experience and understanding of the game.
